Wczytywanie danych

```r
alko <- read.table('http://theta.edu.pl/wp-content/uploads/2018/03/dane_wino.txt', header = TRUE, sep = ',', dec = '.')
head(alko)
```

Jakie są średnie wartości w grupach?

```r
alko1 <- alko[alko$X1 == 1, ]
alko2 <- alko[alko$X1 == 2, ]
alko3 <- alko[alko$X1 == 3, ]

alko1_colmeans <- colMeans(alko1)
alko2_colmeans <- colMeans(alko2)
alko3_colmeans <- colMeans(alko3)

alko1_colmeans
alko2_colmeans
alko3_colmeans
```

Jakie są wartości współczynników funkcji dyskryminacyjnych?

```r
# tworzenie modelu

library(MASS)

jednatrzecia <- 1/3

model_lda = lda(X1 ~ ., data = alko, prior = c(jednatrzecia, jednatrzecia, jednatrzecia))
```

```r
# wyznaczanie współczynników funkcji dyskryminacyjnej

model_lda$scaling
```

Dla każdej z funkcji dyskryminacyjnych określ która ze zmiennych objaśniających jest najbardziej istotna.

```r
significance <- model_lda$scaling


sort(significance[,1], decreasing = TRUE)  # pierwsza zmienna jest najbardziej istotna (X2.8)
sort(significance[,2], decreasing = TRUE) # pierwsza zmienna jest najbardziej istotna (X2.43)
```

Jakie są średnie wartości funkcji dyskryminacyjnych w każdej z grup?

```r
LD = predict(model_lda)$x
#LD
(LD1 = mean(LD[alko$X1 == 1]))   # średnia wartość f.dyskryminacyjnej w grupie 1
(LD2 = mean(LD[alko$X1 == 2]))   # średnia wartość f.dyskryminacyjnej w grupie 2
(LD3 = mean(LD[alko$X1 == 3]))   # średnia wartość f.dyskryminacyjnej w grupie 3
```

Przeprowadź predykcję przynależności do klas na zbiorze testowym

```r
# nie bieżemy zmiennej X1 (klasy) pod uwagę tylko sami przewidujemy, do której klasy dana obserwacja bedzie należeć
# potem zobaczymy jak dobrze przewidzieliśmy

predictions = predict(model_lda)
predictions
```

Jakie są prawdopodobieństwa przynależności do klas dla dwudziestej obserwacji?

```r
predictions$posterior[20,]
```

Do której z klas zostanie zaklasyfikowana ta obserwacja?

```r
predictions$class[20]  # klasa 1
```

Wyznacz procent właściwie zaklasyfikowanych obiektów.

```r
score <- sum(predictions$class == alko$X1) / nrow(alko) * 100
score
```

Przedstaw i skomentuj tabelę predykcji.

```r
(T = table(True = alko$X1, Predicted = predict(model_lda)$class))

# na przekątnej mamy zawsze dobrze zaklasyfikowane obserwacje
# jak widać poza główną przekątną mamy same zero
# więc model zaklasyfikował każdą obserwację do dobrej klasy
```
